```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import json
```

```python
def export_figure (title):
    plt.tight_layout()
    plt.savefig(os.path.join("figs", title + ".png"))
    plt.show()
```

```python
samples = "100"
model_file = "model-women"
results_dir = []
results_file = "cmdstansummary-wg6-seed-1-samples-" + samples + '-model-' + model_file + ".csv"
results = pd.read_csv(os.path.join("model", "results", results_file))
results
```

```python
cohorts = pd.read_csv("model/cohorts_with_ids.csv")
cohorts.sort_values('Students3')
```

```python
enrolments = pd.read_csv("model/enrolments_with_ids.csv")
enrolments.sort_values('Students3').head(20)
```

```python
results.columns = ['variable', 'mean', 'sd', 'c3', 'c97', 'mm', 'ms', 'essb','esst', 'r_hat']
sns.displot(data=results,x='r_hat')
results.sort_values('r_hat', ascending=False).head(40)

```

```python
all_topics = pd.read_csv(os.path.join('curriculum_content','ACM_2023_CAH_codes.csv'))
all_acm_topics = sorted(list(all_topics[all_topics['Area'] != 'Non-CS application']['Abbrev']))
all_cah_topics = sorted(list(set(all_topics['Abbrev']) -  set(all_acm_topics)))
all_cah_topics
```

```python
with open(os.path.join('model','topics.json'), 'rb') as fp:
    topics = json.load(fp)

acm_topics = [topic for topic in topics if topic in all_acm_topics]
cah_topics = [topic for topic in topics if topic in all_cah_topics]
cah_topics
```

```python
# pop_topic_all_means = []
# pop_topic_all_rhats = []
# pop_topic_all_c3 = []
# pop_topic_all_c97 = []

# pop_topic_wm_diff_means = []
# pop_topic_wm_diff_rhats = []
# pop_topic_wm_diff_c3 = []
# pop_topic_wm_diff_c97 = []

pop_topic_women_means = []
pop_topic_women_rhats = []
pop_topic_women_c3 = []
pop_topic_women_c97 = []

pop_topic_men_means = []
pop_topic_men_rhats = []
pop_topic_men_c3 = []
pop_topic_men_c97 = []

for i, topic in enumerate(topics):
    row = results[results['variable'] == "popularity_topic_women[" + str(i) + "]"]
    pop_topic_women_means.append(row.iloc[0]['mean'])
    pop_topic_women_rhats.append(row.iloc[0]['r_hat'])
    pop_topic_women_c3.append(row.iloc[0]['c3'])
    pop_topic_women_c97.append(row.iloc[0]['c97'])

    row = results[results['variable'] == "popularity_topic_men[" + str(i) + "]"]
    pop_topic_men_means.append(row.iloc[0]['mean'])
    pop_topic_men_rhats.append(row.iloc[0]['r_hat'])
    pop_topic_men_c3.append(row.iloc[0]['c3'])
    pop_topic_men_c97.append(row.iloc[0]['c97'])



    
df = pd.DataFrame({"topic": topics, 
                   "pop_topic_women_mean": pop_topic_women_means, 
                   "pop_topic_men_mean": pop_topic_men_means,
                   "pop_topic_women_rhat": pop_topic_women_rhats, 
                   "pop_topic_men_rhat": pop_topic_men_rhats,
                   "pop_topic_women_c3": pop_topic_women_c3,
                   "pop_topic_women_c97": pop_topic_women_c97,
                   "pop_topic_men_c3": pop_topic_men_c3,
                   "pop_topic_men_c97": pop_topic_men_c97,      
                  })


df.sort_values('pop_topic_women_mean', inplace=True)

def plot_with_errors(df, prop, topics, xlabel, ylabel, with_errors=True):
    if prop in ["pop_women", "pop_men"]:
        col = prop
    else:
        col = 'pop_topic_' + prop + '_mean'
    data = df[df['topic'].isin(topics)].sort_values(col)
    y = data[col]
    fig, ax = plt.subplots()
    ax.set_ylim(-800,800)
    ax.axhline(0, color = 'lightgrey', linestyle = 'dashed', linewidth=1)
    ax.plot(data['topic'], y,'ob')
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    
    if with_errors:
        c97 = 'pop_topic_' + prop + '_c97'
        c3 = 'pop_topic_' + prop + '_c3'

        data['upper_error'] = data[c97] - data[col]
        data['lower_error'] = data[col] - data[c3]
        ax.errorbar(data['topic'], y, elinewidth=1, ecolor='black', capsize=3, yerr=[data['lower_error'], data['upper_error']])


    
#    sns.barplot(ax=ax, data=data, x='topic', y='pop_topic_' + prop + '_mean', orient='v')
```

```python
plot_with_errors(df, 'women', acm_topics, "Computing curriculum topics", "Popularity with women")
export_figure("results_acm_women" + samples)
```

```python
plot_with_errors(df, 'men', acm_topics, "Computing curriculum topics", "Popularity with men")
export_figure("results_acm_men" + samples)
```

```python
plot_with_errors(df, 'all', cah_topics, "Non-CS application areas", "Overall popularity")
export_figure("results_cah_all" + samples)
```

```python
plot_with_errors(df, 'wm_diff', cah_topics, "Non-CS application areas", "Popularity difference in women and men")
export_figure("results_cah_wm" + samples)
```
