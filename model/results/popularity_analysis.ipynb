```python
import pandas as pd
import seaborn as sns
```

```python
results_file = "stansummary-wg6-seed-1-samples-2000.csv"
results = pd.read_csv(results_file)
results
```

```python
results.columns = ['variable', 'mean', 'sd', 'c3', 'c97', 'mm', 'ms', 'essb','esst', 'r_hat']
results.head(50)
```

```python
topics = ['1', '10', '12', '13', '14', '15', '16', '17', '18', '19', '2', '20', '21', '22', '3',
 '4', '5', '6', '7', '9', 'AI', 'AL', 'AR', 'DM', 'FPL', 'GIT', 'HCI', 'MSF', 'NC', 'OS',
 'PDC', 'SDF', 'SE', 'SEC', 'SEP', 'SF', 'SPD']

pop_topic_all_means = []
pop_topic_all_rhats = []
pop_topic_wm_diff_means = []
pop_topic_wm_diff_rhats = []

for i, topic in enumerate(topics):
    row = results[results['variable'] == "popularity_topic_all[" + str(i) + "]"]
    pop_topic_all_means.append(row.iloc[0]['mean'])
    pop_topic_all_rhats.append(row.iloc[0]['r_hat'])
    row = results[results['variable'] == "popularity_topic_women_men_diff[" + str(i) + "]"]
    pop_topic_wm_diff_means.append(row.iloc[0]['mean'])
    pop_topic_wm_diff_rhats.append(row.iloc[0]['r_hat'])


    
df = pd.DataFrame({"topic": topics, 
                   "pop_topic_all_mean": pop_topic_all_means, 
                   "pop_topic_wm_diff_mean": pop_topic_wm_diff_means,
                   "pop_topic_all_rhat": pop_topic_all_rhats, 
                   "pop_topic_wm_diff_rhat": pop_topic_wm_diff_rhats
                  })

df.sort_values('pop_topic_all_mean', inplace=True)
sns.barplot(data=df, x='topic', y='pop_topic_all_mean', orient='v')
```

```python
df.sort_values('pop_topic_all_mean')
```

```python
df.sort_values('pop_topic_wm_diff_mean')
```

```python
df['pop_women'] = df['pop_topic_all_mean'] + 0.5*df['pop_topic_wm_diff_mean']
df['pop_men'] = df['pop_topic_all_mean'] - 0.5*df['pop_topic_wm_diff_mean']
```

```python
df.sort_values('pop_women').to_csv("pop_women_men")
```
