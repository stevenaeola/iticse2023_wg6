```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
results_file = "stansummary-wg6-seed-1-samples-5000.csv"
results = pd.read_csv(results_file)
results
```

```python
results.columns = ['variable', 'mean', 'sd', 'c3', 'c97', 'mm', 'ms', 'essb','esst', 'r_hat']
results.head(50)
```

```python
topics = ['1', '10', '12', '13', '14', '15', '16', '17', '18', '19', '2', '20', '21', '22', '3',
 '4', '5', '6', '7', '9', 'AI', 'AL', 'AR', 'DM', 'FPL', 'GIT', 'HCI', 'MSF', 'NC', 'OS',
 'PDC', 'SDF', 'SE', 'SEC', 'SEP', 'SF', 'SPD']

cah_topics = topics[0:20]
acm_topics = topics[20:]
```

```python
sns.__version__
```

```python
pop_topic_all_means = []
pop_topic_all_rhats = []
pop_topic_all_c3 = []
pop_topic_all_c97 = []

pop_topic_wm_diff_means = []
pop_topic_wm_diff_rhats = []
pop_topic_wm_diff_c3 = []
pop_topic_wm_diff_c97 = []



for i, topic in enumerate(topics):
    row = results[results['variable'] == "popularity_topic_all[" + str(i) + "]"]
    pop_topic_all_means.append(row.iloc[0]['mean'])
    pop_topic_all_rhats.append(row.iloc[0]['r_hat'])
    pop_topic_all_c3.append(row.iloc[0]['c3'])
    pop_topic_all_c97.append(row.iloc[0]['c97'])

    row = results[results['variable'] == "popularity_topic_women_men_diff[" + str(i) + "]"]
    pop_topic_wm_diff_means.append(row.iloc[0]['mean'])
    pop_topic_wm_diff_rhats.append(row.iloc[0]['r_hat'])
    pop_topic_wm_diff_c3.append(row.iloc[0]['c3'])
    pop_topic_wm_diff_c97.append(row.iloc[0]['c97'])



    
df = pd.DataFrame({"topic": topics, 
                   "pop_topic_all_mean": pop_topic_all_means, 
                   "pop_topic_wm_diff_mean": pop_topic_wm_diff_means,
                   "pop_topic_all_rhat": pop_topic_all_rhats, 
                   "pop_topic_wm_diff_rhat": pop_topic_wm_diff_rhats,
                   "pop_topic_all_c3": pop_topic_all_c3,
                   "pop_topic_all_c97": pop_topic_all_c97,
                   "pop_topic_wm_diff_c3": pop_topic_wm_diff_c3,
                   "pop_topic_wm_diff_c97": pop_topic_wm_diff_c97,      
                  })

df.sort_values('pop_topic_all_mean', inplace=True)

def plot_with_errors(df, prop, topics):
    col = 'pop_topic_' + prop + '_mean'
    data = df[df['topic'].isin(topics)].sort_values(col)
    col = 'pop_topic_' + prop + '_mean'
    c97 = 'pop_topic_' + prop + '_c97'
    c3 = 'pop_topic_' + prop + '_c3'

    data['upper_error'] = data[c97] - data[col]
    data['lower_error'] = data[col] - data[c3]
    y = data[col]
    fig, ax = plt.subplots()
    ax.errorbar(topics, y, capsize=3, yerr=[data['lower_error'], data['upper_error']])
#    sns.barplot(ax=ax, data=data, x='topic', y='pop_topic_' + prop + '_mean', orient='v')
```

```python
plot_with_errors(df, 'all', acm_topics)
```

```python
plot_with_errors(df, 'wm_diff', acm_topics)
```

```python
plot_with_errors(df, 'all', cah_topics)
```

```python
plot_with_errors(df, 'wm_diff', cah_topics)
```
