```python
#import packages 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
from pandas.plotting import scatter_matrix
from matplotlib.ticker import ScalarFormatter
from matplotlib.ticker import FuncFormatter
from sklearn.preprocessing import MinMaxScaler

#show graphs without the need to involke plt.show()
%matplotlib inline 
```

```python
df=pd.read_csv('result.csv')
df_cah=pd.read_csv('result_CAH.csv')
```

```python
#Plot 1: Overall Popularity for Computing Curriculum Areas

x = df["mean_survey"]
y = df["mean_model"]
t = df["topics"]

r = np.corrcoef(x, y)

fig, ax = plt.subplots()
ax.scatter(x,y,color='grey',edgecolors='black',s=40)

plt.title("Overall Popularity of Computing Curriculum Areas",fontsize=13)
plt.xlabel("Authors' Opinion Survey Mean",fontsize=12)
plt.ylabel("Model Overall Mean",fontsize=12)

for i, txt in enumerate(t):
        if txt == "SEC":        
              ax.annotate(txt, ( x[i]-0.25, y[i]-70) )
        elif txt == "SPD":
              ax.annotate(txt, ( x[i]+0.08, y[i]+20) )
        else:
             ax.annotate(txt, ( x[i]+0.05, y[i]-10) )
                
            
image_format = 'jpg' # e.g .png, .svg, etc.
image_name = 'scatter_plot.jpg'
plt.legend('', frameon=False)
plt.savefig(image_name,dpi=1000)
```

```python
#Plot 2: Overall Popularity for Non-CS Application Areas
x = df_cah["mean_survey"]
y = df_cah["mean_model"]
t = df_cah["topics"]

#Correlation
r_cah = np.corrcoef(x, y)

fig, ax = plt.subplots()
ax.scatter(x,y,color='grey',edgecolors='black',s=40)

plt.legend('', frameon=False)

plt.title("Overall Popularity of Non-CS Application Areas ",fontsize=13)
plt.xlabel("Authors' Opinion Survey Mean",fontsize=12)
plt.ylabel("Model Overall Mean",fontsize=12)

for i, txt in enumerate(t):
        if txt == 7:
              ax.annotate(txt, ( x[i]+0.02, y[i]+5) )
       
        elif txt == 3:
              ax.annotate(txt, ( x[i]-0.07, y[i]) )
        
        elif txt == 2:
              ax.annotate(txt, ( x[i], y[i]-20) )
                
        elif txt == 4:
              ax.annotate(txt, ( x[i], y[i]-20) )
        
        elif txt == 1:
              ax.annotate(txt, ( x[i]+0.03, y[i]-5) )
        
        elif txt == 16:
              ax.annotate(txt, ( x[i]-0.03, y[i]-20) )
        else:
             ax.annotate(txt, ( x[i]-0.03, y[i]+10) )
                
image_format = 'jpg' # e.g .png, .svg, etc.
image_name = 'scatter_plot_cah.jpg'
plt.savefig(image_name,dpi=1000)
```

```python
#Plot 3: Difference in Popularity of Computing Curriculum Areas
x = df["diff_survey"]
y = df["diff_model"]
t = df["topics"]

r_diff = np.corrcoef(x, y)

fig, ax = plt.subplots()
ax.scatter(x,y,color='grey',edgecolors='black',s=40)

plt.title("Difference in Popularity of Computing Curriculum Areas",fontsize=13)
plt.xlabel("Authors' Opinion Survey Mean Difference",fontsize=12)
plt.ylabel("Model Mean Difference",fontsize=12)

for i, txt in enumerate(t):
        if txt == "SF":        
              ax.annotate(txt, ( x[i]-0.3, y[i]) )
        elif txt == "SEP":
              ax.annotate(txt, ( x[i]-0.1, y[i]-15) )
        elif txt == "FPL":
              ax.annotate(txt, ( x[i]-0.1, y[i]-15) )
        else:
             ax.annotate(txt, ( x[i]+0.15, y[i]-5) )
                
            
image_format = 'jpg' # e.g .png, .svg, etc.
image_name = 'scatter_plot_diff.jpg'
plt.legend('', frameon=False)
plt.savefig(image_name,dpi=1000)
```

```python
#Plot 4: Difference in Popularity of Non-CS Application Areas
y = df_cah["diff_model"]
t = df_cah["topics"]

r_diff_cah = np.corrcoef(x, y)

fig, ax = plt.subplots()
ax.scatter(x,y,color='grey',edgecolors='black',s=40)

plt.title("Difference in Popularity of Non-CS Application Areas ",fontsize=13)
plt.xlabel("Authors' Opinion Survey Mean Difference",fontsize=12)
plt.ylabel("Model Mean Difference",fontsize=12)

for i, txt in enumerate(t):
        if txt == 1:
              ax.annotate(txt, ( x[i]+0.1, y[i]) )
       
        elif txt == 19:
              ax.annotate(txt, ( x[i]-0.09, y[i]+0.4) )
        
        elif txt == 2:
              ax.annotate(txt, ( x[i]-0.07, y[i]-1) )
      
        else: 
            ax.annotate(txt, ( x[i]+0.15, y[i]) )
                
            
image_format = 'jpg' # e.g .png, .svg, etc.
image_name = 'scatter_plot_diff_cah.jpg'
plt.legend('', frameon=False)
plt.savefig(image_name,dpi=1000)
```
